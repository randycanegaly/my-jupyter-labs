```python
import plotly.express as px
import pandas as pd

# Create a tiny sample dataset
df = pd.DataFrame({
    "Fruit": ["Apples", "Oranges", "Bananas", "Grapes"],
    "Amount": [4, 1, 2, 8],
    "City": ["SF", "SF", "NYC", "NYC"]
})

# Create an interactive bar chart
fig = px.bar(df, x="Fruit", y="Amount", color="City", barmode="group", title="Interactive Test")
fig.show()
```
